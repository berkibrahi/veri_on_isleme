# eksik veri analizi

```python
import numpy as np
import pandas as pd
v1=np.array([1,3,6 ,np.NaN,7,1,np.NaN,9,15])
v2=np.array([7,np.NaN,5 ,8,12,np.NaN,np.NaN,2,3])
v3=np.array([np.NaN,12,5 ,6,14,7,np.NaN,2,31])
df=pd.DataFrame(
    {
    "v1":v1,
    "v2":v2,
    "v3":v3
})
df
```

```python
#eksik deÄerler
df.isnull().sum()
```

```python
#eksik olmayan deÄerler
df.notnull().sum()
```

```python
#eksik deÄerlerin toplam sayÄ±sÄ±
df.isnull().sum().sum()
```

```python
#eksik deÄer sorgusu
df.isnull()
```

```python
#en az bir eksik deÄerlerin listesi
df[df.isnull().any(axis=1)]
```

```python
# dolu olanlar listesi
df[df.notnull().all(axis=1)]
```

```python
# dolu olanlar listesi alternatif gÃ¶sterimi
df[df["v1"].notnull() & df["v2"].notnull() & df["v3"].notnull()]
```

```python
#eksik deÄerlerin direk siinmesi
df.dropna()
```

```python
#eksik deÄerleri ortalama ile doldurma
df["v1"].fillna(df["v1"].mean())
```

```python
#fonksiyenel programlama kullanÄ±larak eksik deÄerlerin ortalamsÄ± ile doldurulmasÄ±
df.apply(lambda x:x.fillna(x.mean()) ,axis=0)
```
